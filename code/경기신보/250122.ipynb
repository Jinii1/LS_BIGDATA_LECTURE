```python
import pandas as pd
t1 = pd.read_excel('../../data/경기신보/시도별_공장등록_현황.xlsx')
t1
```

```python
## 시각화
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.figure(figsize=(12, 6))

# Clean and prepare the data for visualization
t1['시도'] = t1['시도'].str.strip()  # Remove extra spaces or tabs
t1['합계 비중'] = pd.to_numeric(t1['합계 비중'], errors='coerce')

# Sort by 합계 비중 for better visual representation
t1_sorted = t1.sort_values(by='합계 비중', ascending=False)

# Plotting the 합계 비중 as a bar chart
plt.figure(figsize=(12, 8))
plt.bar(t1_sorted['시도'], t1_sorted['합계 비중'], alpha=0.8, edgecolor='black')

plt.title('시도별 합계 비중', fontsize=16)
plt.xlabel('시도', fontsize=12)
plt.ylabel('합계 비중', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Side-by-side bar chart for 기업 유형 비중 by 시도
x = range(len(t1_sorted['시도']))  # Positions for bars
width = 0.2  # Bar width

plt.figure(figsize=(16, 10))
categories = ['소기업 비중', '중기업 비중', '중견기업 비중', '대기업 비중']

for i, category in enumerate(categories):
    plt.bar(
        [p + i * width for p in x],
        t1_sorted[category],
        width=width,
        label=category,
        alpha=0.8,
        edgecolor='black'
    )

plt.title('시도별 기업 유형 비중', fontsize=16)
plt.xlabel('시도', fontsize=12)
plt.ylabel('비중', fontsize=12)
plt.xticks([p + 1.5 * width for p in x], t1_sorted['시도'], rotation=45, ha='right', fontsize=10)
plt.legend(title='기업 유형', fontsize=12, title_fontsize=12)
plt.tight_layout()
# Display the plot
plt.show()
```
