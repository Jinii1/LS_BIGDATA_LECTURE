# 기존 데이터에서 경기도만 추출하는 코드

```python
gyeonggi_df1 = df1[df1['시도명'] == '경기도']
# gyeonggi_df1.shape # (77787, 38)

# # 시군구명에서 공백 뒤의 내용 삭제
gyeonggi_df1['시군구명'] = gyeonggi_df1['시군구명'].str.split().str[0] # 32

# # 여주군 데이터가 2개가 있는데 관리 기관이 경기도 여주시라서 여주시로 합쳐도 무관하다고 판단
# # '여주군' 데이터를 '여주시' 데이터에 합치기
gyeonggi_df1['시군구명'] = gyeonggi_df1['시군구명'].replace('여주군', '여주시')

# gyeonggi_df1['시군구명'].nunique() # 경기도 행정구역 31개

# ## 엑셀 파일로 저장
gyeonggi_df1.to_csv('../../bigfile/경기_(2024.11월말기준)_전국공장등록현황.csv', encoding="utf-8-sig", index=False)
gyeonggi_df1
```

```python
gyeonggi_df63 = df63[df63['시도명'] == '경기도']

# 시군구명에서 공백 뒤의 내용 삭제
gyeonggi_df63['시군구명'] = gyeonggi_df63['시군구명'].str.split().str[0]
gyeonggi_df63['시군구명'].unique()

# # # 여주군 데이터가 2개가 있는데 관리 기관이 경기도 여주시라서 여주시로 합쳐도 무관하다고 판단
# # # '여주군' 데이터를 '여주시' 데이터에 합치기
gyeonggi_df63['시군구명'] = gyeonggi_df63['시군구명'].replace('여주군', '여주시')

gyeonggi_df63['시군구명'].nunique() # 경기도 행정구역 31개

# 엑셀 파일로 저장
gyeonggi_df63.to_csv('../../bigfile/경기_(2013.12월말기준)_전국공장등록현황.csv', encoding="utf-8-sig", index=False)

# gyeonggi_df63
```

## 경기도 내 시군별로 공장의 입지 구분 (개별입지/계획입지)

```python
# '시군구명'별로 '공장구분'의 분포 확인
factory_by_region = gyeonggi_df1.groupby('시군구명')['공장구분'].value_counts().unstack()
factory_by_region

## 시각화
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.figure(figsize=(9, 5))

# 결측값을 0으로 채우기 (NaN 처리)
factory_by_region_filled = factory_by_region.fillna(0)

# 시각화를 위해 값이 큰 순서로 정렬
factory_by_region_sorted = factory_by_region_filled.sort_values(by='개별입지', ascending=False)

# 막대 그래프 생성
factory_by_region_sorted.plot(kind='bar', figsize=(15, 8), stacked=True)

# 그래프 제목 및 축 레이블 설정
plt.title('2024년 11월 말 기준 경기 시군구별 공장구분 분포', fontsize=16)
plt.xlabel('시군구명', fontsize=12)
plt.ylabel('공장 개수', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.legend(title='공장구분', fontsize=10)
plt.tight_layout()

# 그래프 표시
plt.show()

# 화성시와 김포시의 압도적인 공장 수 -> 제조업 및 공업 활동이 활발한 지역? 그런가봄
# 화성시와 김포시는 개별입지 공장이 많고, 안산시는 계획입지 공장이 지배적
# 개별입지: 개별적으로 공장 설립, 계획입지: 계획된 공업단지 내에 설립된 공장 의미
```

### 거기에 따른 종업원수 규모같은걸 보고 싶다면?

```python
# 시군구별 종업원 수 합계와 평균 계산 (각 시군별 공장에서 종업원합계 값의 평균 = 시군별 공장 한 곳당 평균적으로 몇 명의 종업원이 있는지)
employees_by_region = gyeonggi_df1.groupby('시군구명')['종업원합계'].agg(['sum', 'mean']).fillna(0)
# sum: 모든 공장의 종업원 수 합산, mean: 공장별 종업원합계의 평균값 (총합/시군별 공장 갯수)
employees_by_region = employees_by_region.sort_values(by='sum', ascending=False)
print(employees_by_region)

# 평택시: 평균이 높음 (공장당 종업원 수가 매우 높음) -> 대규모 공장일까? 대기업이나 산업단지와 관련이 있을까?

# 수원시, 오산시: 공장이 많진 않지만 공장 당 종업원 수 매우 높음 -> 대규모 공장일까?

# 양평군: 총합과 평균 모두 작음, 제조업보단 다른 산업이 주를 이룰까?
```

평택시: 평균이 높음 (공장당 종업원 수가 매우 높음)
-> 대규모 공장, 대기업이거나 산업단지와 관련이 있을 것

시군구별 공장규모 분포 확인

```python
factory_size_distribution = gyeonggi_df1.groupby(['시군구명', '공장규모']).size().unstack(fill_value=0)

# 공장 규모 순서를 재정렬 (대기업 > 중견기업 > 중기업 > 소기업 순)
factory_size_distribution = factory_size_distribution[['대기업', '중견기업', '중기업', '소기업']]

# 대기업 기준으로 내림차순 정렬
factory_size_distribution_sorted = factory_size_distribution.sort_values(by='대기업', ascending=False)
factory_size_distribution_sorted
```
