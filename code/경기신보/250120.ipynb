```python
import pandas as pd
t1 = pd.read_excel('../../data/경기신보/시도별_공장등록_현황.xlsx')
t1
t1_filtered = t1[t1['시도'] != '합계']
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.figure(figsize=(9, 5))

regions = t1_filtered['시도']
proportions = t1_filtered['합계 비중']

plt.figure(figsize=(8, 8))
plt.pie(proportions, labels=regions, autopct='%1.1f%%', startangle=90, counterclock=False)
plt.title('시도별 합계 비중')
plt.show()







```

```python
# Adjusting the pie chart to improve readability by adding an explode effect and ensuring labels are properly spaced

# Adding an explode effect for better visibility of slices
explode = [0.05] * len(proportions)  # Slightly separate all slices

# Creating a more readable pie chart
plt.figure(figsize=(10, 10))
plt.pie(
    proportions,
    labels=regions,
    autopct='%1.1f%%',
    startangle=90,
    counterclock=False,
    explode=explode,
    pctdistance=0.85,  # Position percentage values closer to slices
    labeldistance=1.05,  # Position labels slightly further from slices
)
plt.title('시도별 합계 비중')
plt.show()
```

```python
# Bar Plot: 시도별 공장 등록 개수
plt.figure(figsize=(12, 6))
plt.bar(t1_filtered['시도'], t1_filtered['합계 개사'], color='skyblue')
plt.title('시도별 공장 등록 개수', fontsize=16)
plt.xlabel('시도', fontsize=12)
plt.ylabel('공장 등록 개수', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Pie Chart: 시도별 공장 등록 비중
plt.figure(figsize=(10, 10))
plt.pie(t1_filtered['합계 비중'], labels=t1_filtered['시도'], autopct='%1.1f%%', startangle=90, counterclock=False, colors=plt.cm.tab20.colors)
plt.title('시도별 공장 등록 비중', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Horizontal Bar Chart: 시도별 공장 등록 비중 (sorted)
sido_df_sorted = t1_filtered.sort_values(by='합계 비중', ascending=True)
plt.figure(figsize=(10, 6))
plt.barh(sido_df_sorted['시도'], sido_df_sorted['합계 비중'], color='lightgreen')
plt.title('시도별 공장 등록 비중 (정렬)', fontsize=16)
plt.xlabel('비중 (%)', fontsize=12)
plt.ylabel('시도', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
t6 = pd.read_excel('../../data/경기신보/고용현황.xlsx')
t6
```

```python
plt.figure(figsize=(12, 6))
plt.bar(t6["시도"], t6["내국인 남자"], label="내국인 남자", color='lightblue')
plt.bar(t6["시도"], t6["내국인 여자"], bottom=t6["내국인 남자"], label="내국인 여자", color='pink')

plt.title("시도별 내국인 고용현황 (남자와 여자)")
plt.ylabel("고용 소계")
plt.xlabel("시도")
plt.xticks(rotation=45)
plt.ylim(0, 1300000)
plt.ticklabel_format(style='plain', axis='y')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))

# Plot bars
plt.bar(t6["시도"], t6["내국인 남자"], label="내국인 남자", color='lightblue')
plt.bar(t6["시도"], t6["내국인 여자"], 
        bottom=t6["내국인 남자"], label="내국인 여자", color='pink')

# Add text for "내국인 여성비율"
for idx, row in t6.iterrows():
    female_ratio = row["내국인 여자"] / (row["내국인 남자"] + row["내국인 여자"]) * 100
    plt.text(idx, row["내국인 남자"] + row["내국인 여자"] / 2, 
             f'{female_ratio:.1f}%', ha='center', fontsize=9, color='black')

# Formatting the chart
plt.title("시도별 내국인 고용현황 (남자와 여자)")
plt.ylabel("고용 소계")
plt.xlabel("시도")
plt.xticks(rotation=45)
plt.ylim(0, 1300000)
plt.ticklabel_format(style='plain', axis='y')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Re-plot the chart without foreigner percentage text
plt.figure(figsize=(12, 6))

# Plot bars for Korean male, Korean female, and Foreigners
plt.bar(t6["시도"], t6["내국인 남자"], label="내국인 남자", color='lightblue')
plt.bar(t6["시도"], t6["내국인 여자"], 
        bottom=t6["내국인 남자"], label="내국인 여자", color='pink')
plt.bar(t6["시도"], t6["외국인"], 
        bottom=t6["내국인 남자"] + t6["내국인 여자"], 
        label="외국인", color='orange')

# Add text for "내국인 여성비율" only
for idx, row in t6.iterrows():
    # Calculate position for Korean female ratio
    female_ratio = row["내국인 여자"] / (row["내국인 남자"] + row["내국인 여자"]) * 100
    
    # Add text for Korean female ratio in the middle of the female bar
    plt.text(idx, 
             row["내국인 남자"] + (row["내국인 여자"] / 2), 
             f'{female_ratio:.1f}%', 
             ha='center', 
             fontsize=8, 
             color='black', 
             rotation=0)

# Formatting the chart
plt.title("시도별 내국인 및 외국인 고용현황 (남자, 여자, 외국인)")
plt.ylabel("고용 소계")
plt.xlabel("시도")
plt.xticks(rotation=45)
plt.ylim(0, 1400000)  # Adjust y-axis limit to fit the bars
plt.ticklabel_format(style='plain', axis='y')  # Use plain number format for y-axis
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import plotly.express as px

# Transform the data for visualization
melted_df = t7.melt(id_vars=['시도'], var_name='업종', value_name='사업체 수')

# Create a treemap with improved readability
fig = px.treemap(
    melted_df,
    path=['시도', '업종'],  # Hierarchical levels: 시도 -> 업종
    values='사업체 수',     # Size of each block
    title="시도별 업종별 트리맵",
    color='사업체 수',     # Color scale based on the business count
    color_continuous_scale='Blues',  # A visually distinct color palette
    labels={'사업체 수': '사업체 수'},  # Ensure labels are user-friendly
)

# Adjust layout and font sizes for better readability
fig.update_layout(
    title_font_size=24,  # Larger title font size
    title_x=0.5,  # Center align the title
    margin=dict(t=50, l=50, r=50, b=50),  # Slightly increased margins
    font=dict(size=14),  # Larger overall font size
)

# Adjust treemap text font sizes specifically
fig.update_traces(
    textinfo="label+value",  # Show both label and value
    textfont=dict(size=16),  # Larger font size for block labels
)

fig.show()
```

```python
import matplotlib.pyplot as plt

# Filter the data for a specific 시도 (e.g., 경기도)
selected_sido = "경기"
sido_data = t7[t7['시도'] == selected_sido].iloc[0, 1:]  # Exclude the 시도 column

# Sort data to find the top 5 업종 by 사업체 수
top5_sido_data = sido_data.sort_values(ascending=False)[:5]

# Create a horizontal bar chart for the top 5 업종
plt.figure(figsize=(10, 6))
top5_sido_data.plot(kind='barh', color='skyblue', edgecolor='black')
plt.xlabel("사업체 수", fontsize=14)
plt.ylabel("업종", fontsize=14)
plt.title(f"{selected_sido} 주요 업종 (Top 5)", fontsize=16)
plt.gca().invert_yaxis()  # Invert y-axis to show the highest value at the top
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Prepare data for Facet Grid
melted_df = t7.melt(id_vars=['시도'], var_name='업종', value_name='사업체 수')

# Create a Facet Grid by 시도
g = sns.FacetGrid(melted_df, col='시도', col_wrap=4, height=4, sharey=False)
g.map(sns.barplot, '업종', '사업체 수', order=melted_df['업종'].unique(), palette='muted')
g.set_xticklabels(rotation=90)
g.set_titles("{col_name}")
plt.subplots_adjust(top=0.9)
g.fig.suptitle('시도별 업종 사업체 수 분포', fontsize=16)
plt.show()
```

```python
import plotly.express as px

# Simulated longitude and latitude for 시도
location_data = {
    "시도": ["경기", "경남", "경북", "인천", "충남"],
    "위도": [37.4363177, 35.4414209, 36.6308397, 37.4559418, 36.6173379],
    "경도": [127.550802, 128.2417453, 128.962578, 126.7051505, 126.8453965]
}
location_df = pd.DataFrame(location_data)

# Aggregate 업종별 데이터 by 시도
aggregated_data = t7.drop('시도', axis=1).sum(axis=1)
location_df['총 사업체 수'] = aggregated_data

# Create map
fig = px.scatter_geo(
    location_df,
    lat='위도',
    lon='경도',
    size='총 사업체 수',
    hover_name='시도',
    title='시도별 사업체 분포 지도',
    template='plotly_white'
)
fig.update_geos(
    projection_type="natural earth",
    center={"lat": 36.5, "lon": 127.8},  # Center of South Korea
    projection_scale=20,  # Zoom level
    visible=True
)
fig.show()
```
