### 시도별 공장등록 면적 현황

```python
t3 = pd.read_excel('../../data/경기신보/시도별_공장등록_면적_현황.xlsx')
t3
t3_filtered = t3[t3['시도'] != '합계']
t3_filtered
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.figure(figsize=(9, 5))


# 시각화를 위해 데이터 설정
labels = t3_filtered['시도']
below_500 = t3_filtered['500㎡미만 소계']
above_500 = t3_filtered['500㎡이상 소계']

# 막대그래프 생성
x = range(len(labels))
width = 0.4

# 500㎡ 미만 소계 그래프
plt.bar([i - width / 2 for i in x], below_500, width, label='500㎡ 미만', color='skyblue')

# 500㎡ 이상 소계 그래프
plt.bar([i + width / 2 for i in x], above_500, width, label='500㎡ 이상', color='pink')

# 그래프 설정
plt.xlabel('시도', fontsize=12)
plt.ylabel('공장 등록 수', fontsize=12)
plt.title('시도별 공장등록 면적 현황', fontsize=14)
plt.xticks(x, labels, rotation=45)
plt.legend(title_fontsize=14, fontsize=12)


# 그래프 출력
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 준비
categories = ['~199㎡', '~499㎡', '~999㎡', '~4,999㎡', '~9,999㎡', '10,000~㎡']
data = t3_filtered[categories]
labels = t3_filtered['시도']

# 색상 설정
colors = ['skyblue', 'lightgreen', 'orange', 'pink', 'purple', 'gray']

# 스택형 막대그래프 생성
x = np.arange(len(labels))  # x축 위치
width = 0.8  # 막대 너비

# 그래프 초기화
fig, ax = plt.subplots(figsize=(12, 8))

# 누적 값 초기화
bottom_values = np.zeros(len(labels))

# 각 카테고리 데이터를 스택으로 쌓아 그래프 생성
for idx, category in enumerate(categories):
    ax.bar(x, data[category], width, label=category, color=colors[idx], bottom=bottom_values)
    bottom_values += data[category]

# 그래프 설정
ax.set_xlabel('시도', fontsize=12)
ax.set_ylabel('공장 등록 수', fontsize=12)
ax.set_title('시도별 공장 크기별 분포', fontsize=14)
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=45)
ax.legend(title='공장 크기', title_fontsize=14, fontsize=12) 

# 그래프 출력
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 100% 스택형 막대그래프 데이터 설정
labels = t3_filtered['시도']
below_500_ratio = t3_filtered['500㎡미만 비중']
above_500_ratio = t3_filtered['500㎡이상 비중']

# x축 위치 설정
x = np.arange(len(labels))
width = 0.8

# 그래프 초기화
fig, ax = plt.subplots(figsize=(12, 8))

# 500㎡ 미만 비중 그래프
ax.bar(x, below_500_ratio, width, label='500㎡ 미만', color='skyblue')

# 500㎡ 이상 비중 그래프
ax.bar(x, above_500_ratio, width, bottom=below_500_ratio, label='500㎡ 이상', color='pink')

# 그래프 설정
ax.set_xlabel('시도', fontsize=12)
ax.set_ylabel('비중', fontsize=12)
ax.set_title('시도별 500㎡ 미만 및 500㎡ 이상 비중 비교', fontsize=14)
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=45)
ax.legend(title='공장 크기')

# 비율 값 표시
for i in range(len(labels)):
    ax.text(x[i], below_500_ratio[i] / 2, f"{below_500_ratio[i]:.1%}", ha='center', va='center', color='black', fontsize=10)
    ax.text(x[i], below_500_ratio[i] + above_500_ratio[i] / 2, f"{above_500_ratio[i]:.1%}", ha='center', va='center', color='black', fontsize=10)

# 그래프 출력
plt.tight_layout()
plt.show()
```

### 용도지역별 공장등록 현황

```python
t5 = pd.read_excel('../../data/경기신보/용도지역별_공장등록_현황.xlsx')
t5= t5[t5['시도'] != '합계']
t5
```

```python
# Extract necessary columns for the bar chart
categories = ['도시지역 개사', '관리지역 개사', '농림지역 개사', '자연지역 개사']
data_for_plot = t5[['시도'] + categories]

labels = data_for_plot['시도']
data_values = data_for_plot[categories]

# Stacked bar chart settings
x = np.arange(len(labels))  # x-axis positions
colors = ['skyblue', 'pink', 'green', 'purple']

# Plot the stacked bar chart
fig, ax = plt.subplots(figsize=(12, 8))
bottom_values = np.zeros(len(labels))

for idx, category in enumerate(categories):
    ax.bar(x, data_values[category], label=category, color=colors[idx], bottom=bottom_values)
    bottom_values += data_values[category]

# Customize the chart
ax.set_xlabel('시도', fontsize=12)
ax.set_ylabel('공장 등록 개수', fontsize=12)
ax.set_title('용도지역별 공장등록 현황', fontsize=14)
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=45)
ax.legend(title='용도지역', fontsize=12, title_fontsize=12)

# Adjust layout and display
plt.tight_layout()
plt.show()
```

```python
t7 = pd.read_excel('../../data/경기신보/업종현황.xlsx')
t7_filtered = t7[t7['시도'] != '합계']

# Transform data to plot industries on the x-axis and group by region
categories = t7_filtered.columns[1:]  # Extract all industry columns
data_for_plot = t7_filtered.set_index('시도').T  # Transpose data for industries on x-axis

# Plot grouped bar chart
fig, ax = plt.subplots(figsize=(14, 8))
x = np.arange(len(categories))  # x positions
width = 0.15  # Bar width

# Plot each region as a group of bars
for i, region in enumerate(data_for_plot.columns):
    ax.bar(x + i * width, data_for_plot[region], width, label=region)

# Customize the chart
ax.set_xlabel('업종', fontsize=12)
ax.set_ylabel('갯수', fontsize=12)
ax.set_title('업종별 지역별 공장 등록 현황', fontsize=14)
ax.set_xticks(x + width * (len(data_for_plot.columns) - 1) / 2)
ax.set_xticklabels(categories, rotation=45)
ax.legend(title='지역', fontsize=10, title_fontsize=12)

# Adjust layout and display
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# 데이터 준비 (업종별 공장 등록 수 합계 계산)
industry_totals = t7_filtered.iloc[:, 1:].sum(axis=0)  # 업종별 총합 계산
industries = industry_totals.index  # 업종명
totals = industry_totals.values  # 총 공장 등록 수

# 원형 아이콘 크기 설정 (공장 등록 수를 반영)
sizes = totals / max(totals) * 1000  # 크기를 공장 등록 수에 비례하게 설정

# 그래프 생성
fig, ax = plt.subplots(figsize=(12, 8))

# 원형 아이콘 플롯 생성
for i, (industry, size) in enumerate(zip(industries, sizes)):
    ax.scatter(i, 0, s=size, alpha=0.6, label=industry)

# 아이콘 내부에 공장 등록 수 표시
for i, total in enumerate(totals):
    ax.text(i, 0, f'{total:,}', ha='center', va='center', fontsize=10, color='black')

# 그래프 설정
ax.set_xlim(-1, len(industries))  # x축 범위 조정
ax.set_xticks(range(len(industries)))  # 업종 이름 표시를 위한 x축 눈금
ax.set_xticklabels(industries, rotation=45, ha='right', fontsize=10)
ax.set_yticks([])  # y축 제거 (포커스는 원형 아이콘)
ax.set_title('업종별 주요 산업군 포커스', fontsize=14)

# 여백 조정 및 출력
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
import matplotlib.pyplot as plt


# 그래프 크기 조정
fig, ax = plt.subplots(figsize=(12, 4))  # 높이를 줄임

# 원형 아이콘 재생성
for i, (industry, size) in enumerate(zip(industries, sizes)):
    ax.scatter(i, 0, s=size, alpha=0.6, label=industry)

# 원 내부에 값 추가
for i, total in enumerate(totals):
    ax.text(i, 0, f'{total:,}', ha='center', va='center', fontsize=10, color='black')

# x축 설정
ax.set_xlim(-1, len(industries))
ax.set_xticks(range(len(industries)))
ax.set_xticklabels(industries, rotation=45, ha='right', fontsize=10)

# y축 제거
ax.set_yticks([])

# 제목 설정
ax.set_title('업종별 주요 산업군 포커스', fontsize=14)

# 레이아웃 조정
plt.tight_layout()
plt.show()
```

```python
# 그래프: 드라마틱과 기본 사이 버전
fig, ax = plt.subplots(figsize=(12, 5))  # 중간 크기 설정

# 원 크기 조정 (드라마틱과 기본 사이)
scaled_sizes = sizes ** 1.2  # 크기 조정

# 색상 설정
cmap = plt.cm.get_cmap('spring', len(industries))  # 색상 팔레트 변경

# 원형 아이콘 생성
for i, (industry, size) in enumerate(zip(industries, scaled_sizes)):
    color = cmap(i / len(industries))
    ax.scatter(i, 0, s=size, alpha=0.7, color=color, edgecolors='white', linewidth=0.8)

# 값 추가 (원 중심에 숫자)
for i, total in enumerate(totals):
    ax.text(i, 0, f'{total:,}', ha='center', va='center', fontsize=10, color='black', fontweight='medium')

# x축 설정
ax.set_xlim(-1, len(industries))
ax.set_xticks(range(len(industries)))
ax.set_xticklabels(industries, rotation=45, ha='right', fontsize=9)
ax.set_yticks([])  # y축 제거
ax.set_title('업종별 주요 산업군 포커스', fontsize=14)

# 배경 설정
ax.set_facecolor('#f0f0f0')  # 배경색을 부드러운 회색으로 설정
plt.tight_layout()
plt.show()
```
