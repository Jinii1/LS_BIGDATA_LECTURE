### 시군구별 건축면적 (제조시설+부대시설)

```python
import numpy as np

conditions = [
    (gyeonggi_df1['건축면적'] <200),
    (gyeonggi_df1['건축면적'] >= 200) & (gyeonggi_df1['건축면적'] <500),
    (gyeonggi_df1['건축면적'] >= 500) & (gyeonggi_df1['건축면적'] <1000),
    (gyeonggi_df1['건축면적'] >= 1000) & (gyeonggi_df1['건축면적'] <5000),
    (gyeonggi_df1['건축면적'] >= 5000) & (gyeonggi_df1['건축면적'] <10000),
    (gyeonggi_df1['건축면적'] >= 10000)
]

labels = ['~199㎡', '200~499㎡', '500~999㎡', '1000~4999㎡', '5000~9999㎡', '10000㎡ ~']

gyeonggi_df1['건축면적_범위'] = np.select(conditions, labels, default='Unknown')

result = gyeonggi_df1.groupby(['시군구명', '건축면적_범위']).size().unstack(fill_value=0)
result
result.to_excel('../../bigfile/시군구별_건축면적.xlsx')
```

### 규모별 공장등록 면적 현황

```python
gyeonggi_df1

required_columns = ['공장규모', '용지면적', '제조시설면적', '부대시설면적']
df_filtered = gyeonggi_df1[required_columns]

# 시군구별 합계 계산
df_grouped = df_filtered.groupby(['공장규모']).sum().reset_index()
df_grouped

# 전체 합계를 기준으로 비율 계산
total_land = df_grouped['용지면적'].sum()
total_manufacture = df_grouped['제조시설면적'].sum()
total_facility = df_grouped['부대시설면적'].sum()

df_grouped['용지면적_비율'] = (df_grouped['용지면적'] / total_land * 100).round(4)
df_grouped['제조시설면적_비율'] = (df_grouped['제조시설면적'] / total_manufacture * 100).round(4)
df_grouped['부대시설면적_비율'] = (df_grouped['부대시설면적'] / total_facility * 100).round(4)

df_grouped[['용지면적', '제조시설면적', '부대시설면적']] = df_grouped[['용지면적', '제조시설면적', '부대시설면적']].astype(int)
df_grouped

df_grouped.to_csv('../../bigfile/규모별면적.csv', index=False)
```
