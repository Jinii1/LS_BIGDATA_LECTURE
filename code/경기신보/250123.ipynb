```python
t2 = pd.read_excel('../../data/경기신보/시도_권역별_공장등록_변동_추이.xlsx')
t2
```

```python
## 시각화
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.figure(figsize=(9, 5))

# Extracting relevant columns for the stack area graph
years = t2['연도']
regions = {
    '수도권': t2['수도권 공장'],
    '부울경': t2['부울경 공장'],
    '충청권': t2['충청권 공장'],
    '호남권': t2['호남권 공장'],
    '대경권': t2['대경권 공장'],
    '강원·제주': t2['강원 제주 공장']
}

# Plotting the stack area graph
plt.figure(figsize=(12, 6))
plt.stackplot(years, regions.values(), labels=regions.keys(), alpha=0.8)
plt.title('권역별 등록 공장 수 변동 추이 (2006년~2024년)', fontsize=14)
plt.xlabel('연도', fontsize=12)
plt.ylabel('공장 수', fontsize=12)
plt.legend(loc='upper left', fontsize=10)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
## 시각화
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.figure(figsize=(9, 5))

# Filter data for 2024
data_2024 = t2[t2['연도'] == '2024년']

# Extract regions and their proportions for the pie chart
regions = ['수도권', '부울경', '충청권', '호남권', '대경권', '강원 제주']
proportions = [
    data_2024['수도권 비율'].values[0],
    data_2024['부울경 비율'].values[0],
    data_2024['충청권 비율'].values[0],
    data_2024['호남권 비율'].values[0],
    data_2024['대경권 비율'].values[0],
    data_2024['강원 제주 비율'].values[0],
]


# Define pastel colors for the pie chart
pastel_colors = ['#FFB3BA', '#FFDFBA', '#FFFFBA', '#BAFFC9', '#BAE1FF', '#D5BAFF']

# Re-plot the pie chart with pastel colors
plt.figure(figsize=(8, 8))
plt.pie(proportions, labels=regions, autopct='%.1f%%', startangle=90, colors=pastel_colors)
plt.title('2024년 권역별 공장등록 비중', fontsize=14)
plt.show()
```

```python
t4 = pd.read_excel('../../data/경기신보/시도_권역별_미만_면적.xlsx')
t4
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data preparation for 누적 막대그래프
regions = ['수도권 공장', '부울경 공장', '충청권 공장', '호남권 공장', '대경권 공장', '강원 제주 공장']
stack_data = t4[regions].T
years = t4['연도']

# 누적 막대그래프: 업종별 공장 등록 수를 시도별로 나누어 시각화
plt.figure(figsize=(12, 6))
stack_data.index = regions
plt.bar(years, stack_data.iloc[0], label=regions[0])
for i in range(1, len(regions)):
    plt.bar(years, stack_data.iloc[i], bottom=stack_data.iloc[:i].sum(), label=regions[i])

# Formatting
plt.title('업종별 공장 등록 수 누적 그래프', fontsize=14)
plt.xlabel('연도', fontsize=12)
plt.ylabel('공장 수', fontsize=12)
plt.xticks(rotation=45)
plt.legend(loc='upper left', fontsize=10)
plt.tight_layout()
plt.show()
```

```python
t8 = pd.read_excel('../../data/경기신보/업종별_공장등록_면적.xlsx')
```

```python
# 데이터 준비
industries = t8['업종']
average_areas = t8['평균제조면적']
total_areas = t8['합계면적']

# 막대그래프: 업종별 평균 제조 면적
plt.figure(figsize=(10, 6))
plt.bar(industries, average_areas, color='skyblue')
plt.xticks(rotation=90)
plt.title('업종별 평균 제조 면적', fontsize=14)
plt.xlabel('업종', fontsize=12)
plt.ylabel('평균 제조 면적 (㎡)', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# 업종별 평균 제조 면적과 공장 등록수를 선 그래프로 추가하여 시각화
fig, ax1 = plt.subplots(figsize=(12, 6))

# 막대그래프: 평균 제조 면적
bars = ax1.bar(industries, average_areas, color='skyblue', label='평균 제조 면적 (㎡)')
ax1.set_xlabel('업종', fontsize=12)
ax1.set_ylabel('평균 제조 면적 (㎡)', fontsize=12, color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.set_xticklabels(industries, rotation=90)
ax1.legend(loc='upper left')

# 선 그래프: 공장 등록 수
ax2 = ax1.twinx()
ax2.plot(industries, t8['공장등록수'], color='red', marker='o', label='공장 등록 수')
ax2.set_ylabel('공장 등록 수', fontsize=12, color='red')
ax2.tick_params(axis='y', labelcolor='red')
ax2.legend(loc='upper right')

plt.title('업종별 평균 제조 면적 및 공장 등록 수', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
t6 = pd.read_excel('../../data/경기신보/고용현황.xlsx')
```

```python
import matplotlib.pyplot as plt

# Data preparation for 다중 막대그래프
regions = t6['시도']
domestic_ratios = t6['내국인 비중']
foreign_ratios = t6['외국인 비중']

x = range(len(regions))
bar_width = 0.4

# 다중 막대그래프: 내국인 vs 외국인 고용 비율
plt.figure(figsize=(12, 6))
plt.bar(x, domestic_ratios, width=bar_width, label='내국인 비중', color='skyblue')
plt.bar([p + bar_width for p in x], foreign_ratios, width=bar_width, label='외국인 비중', color='orange')

# Formatting
plt.xlabel('지역', fontsize=12)
plt.ylabel('고용 비율', fontsize=12)
plt.title('지역별 내국인과 외국인 고용 비율 비교', fontsize=14)
plt.xticks([p + bar_width / 2 for p in x], regions, rotation=90)
plt.legend()
plt.tight_layout()
plt.show()

# # Data preparation for 버블차트
# total_employment = t6['전체 합계']
# foreign_ratio = t6['외국인 비율']
# bubble_sizes = total_employment / 1000  # Scale for bubble sizes

# # 버블차트: 고용 규모와 외국인 비율
# plt.figure(figsize=(10, 6))
# plt.scatter(foreign_ratio, total_employment, s=bubble_sizes, alpha=0.6, color='purple')

# # Formatting
# plt.xlabel('외국인 비율', fontsize=12)
# plt.ylabel('전체 고용 규모', fontsize=12)
# plt.title('지역별 고용 규모와 외국인 비율', fontsize=14)

# # Annotating regions
# for i, txt in enumerate(regions):
#     plt.annotate(txt, (foreign_ratio[i], total_employment[i]), fontsize=10, ha='right')

# plt.tight_layout()
# plt.show()
```

```python
# Data preparation for 버블차트
total_employment = t6['전체 합계']
foreign_ratio = t6['외국인 비율']
bubble_sizes = total_employment / 1000  # Scale for bubble sizes

# 버블차트: 고용 규모와 외국인 비율
plt.figure(figsize=(10, 6))
plt.scatter(foreign_ratio, total_employment, s=bubble_sizes, alpha=0.6, color='purple')

# Formatting
plt.xlabel('외국인 비율', fontsize=12)
plt.ylabel('전체 고용 규모', fontsize=12)
plt.title('지역별 고용 규모와 외국인 비율', fontsize=14)

# Annotating regions
for i, txt in enumerate(regions):
    plt.annotate(txt, (foreign_ratio[i], total_employment[i]), fontsize=10, ha='right')

plt.tight_layout()
plt.show()
```
